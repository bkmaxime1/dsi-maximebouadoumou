We would like to use a regression tree to predict the score of a wine. In order to do that, we first need to select and engineer appropriate features.

```python
#with all the scores that are not NAN --find all the null that not using ~means NOT missing
#create new dataframe without null for score
data =data[~data["Score"].isnull()]
#delete all score above 10
data =data[data["Score"]<=10]
data
```

```python
#create some dummies on meaningful things
import patsy
y,X=patsy.dmatrices("Score ~ C(Country)+ C(Grape)+ C(Region) + Price + C(Vintage) + C(Vinyard)",data)
```

```python
from sklearn.cross_validation import train_test_split
# from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

dtr =DecisionTreeRegressor()
model =dtr.fit(X_train,y_train)
predictions = model.predict(X_test)

predictions
```
