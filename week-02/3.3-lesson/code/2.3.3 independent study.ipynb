```python
import pandas as pd
```

```python
#read data in the raw data from github url

#df = pd.read_csv('https://raw.githubusercontent.com/suneel0101/lesson-plan/master/crunchbase_monthly_export.csv')
#replace na by empty
df = pd.read_csv('https://raw.githubusercontent.com/suneel0101/lesson-plan/master/crunchbase_monthly_export.csv',na_values = [' _ ', "na","*"])
```

```python
#find missing value
# df.isnull()
# df.notnull()
df['funding_rounds'].isnull().sum()# do the sum of missing values if exists
df['funding_rounds'].isnull().unique()# check for unique numbers
df['funding_rounds'].isnull().values.any()
```

```python
#filling in missing data with the mean
df[' funding_total_usd '] = df[' funding_total_usd '].str.replace(',','')
df[' funding_total_usd '] = df[' funding_total_usd '].str.replace(' ','')
```

```python
Convert the data type to float
df[' funding_total_usd '] = df[' funding_total_usd '].astype(float) 
```

```python
df[' funding_total_usd '] = pd.to_numeric(df[' funding_total_usd '], errors = "coerce" 
df
```

```python
df[' funding_total_usd '] = df[' funding_total_usd '].fillna(df[' funding_total_usd '].mean(skipna=True))
df
```

```python
df[' funding_total_usd '] = df[' funding_total_usd '].astype(int)
df
```

```python
#drop 3 columns
df.dropna(subset=['founded_year', 'founded_month','state_code'], inplace = True)
df
```
